```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

# Enter The registered name of your fav stock

```python
company = input("Enter your company: ")
```

```python
comp = yf.Ticker(company)
```

```python
hist = comp.history(period="max")
```

```python
a = hist.to_numpy()
shape = a.shape
end = shape[0]
```

```python
open = []
for i in range(0, end):
    open.append(a[i][0])
    
close = []
for i in range(0,end):
    close.append(a[i][3])
    
High = []
for i in range(0,end):
    High.append(a[i][1])
    
Low = []
for i in range(0,end):
    Low.append(a[i][2])
    
volume = []
for i in range(0,end):
    volume.append(a[i][4])
    
    
count = []
for i in range(0,end):
    a = 0
    a = a + i
    count.append(a)
```

```python
plt.plot(open, label='open', color ='green', linestyle = '-', linewidth =1.5)
plt.plot(close, label='close', color ='red', linestyle = '--', linewidth =1.5)
plt.legend()
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Opening & Closing Stock Prices')
plt.show()
```

```python
plt.plot(High, label='High', color ='green', linestyle = '-', linewidth =1.5)
plt.plot(Low, label='Low', color ='red', linestyle = '--', linewidth =1.5)
plt.legend()
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('High & Low Stock Prices')
plt.show()
```

```python
plt.bar(count,volume)
plt.xlabel('Days')
plt.ylabel('Volume')
plt.show()
```

# Your Portfolio

```python
num = int(input("Enter the number of stocks you want to invest in: "))
```

```python
list = []
invest = []
bought = []
qtys = []
for i in range(0,num):
    a = input("Enter the company name: ")
    list.append(a)
    qty = int(input("Enter the Quantity: "))
    qtys.append(qty)
    price = int(input("Enter the price of the share that day: "))
    bought.append(price)
    investment = qty*price
    invest.append(investment)
```

```python
plt.pie(invest, labels = list, autopct ="%1.1f%%")
```

```python
latest = []
for i in range(0,num):
    comp = yf.Ticker(list[i])
    hist = comp.history(period="max")
    a = hist.to_numpy()
    shape = a.shape
    end = shape[0]
    latest.append(a[end-1][3])
```

```python
X_axis = np.arange(len(list))
  
plt.bar(X_axis - 0.2, bought, 0.4, label = 'Investment')
plt.bar(X_axis + 0.2, latest, 0.4, label = 'Current Price')
  
plt.xticks(X_axis, list)
plt.xlabel("Companies")
plt.ylabel("Price")
plt.title("Price Comparision")
plt.legend()
plt.show()
```
